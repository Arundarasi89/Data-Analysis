```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = d.corr()
sns.heatmap(corr)
```

```python
plt.hist(d['age'],color='red',histtype='bar',rwidth=1.0)
plt.xlabel("patient age")
plt.ylabel("Frequency")
plt.title("Histogram Plot: Age of patients with chronic kidney disease")
plt.show()
```

```python
x1 = d['pcc'].value_counts()
```

```python
x1
```

```python
plt.figure(figsize=(5,5))
label = ['present','notpresent']
c=['b','r']
plt.pie(x1, labels=label,colors=c,startangle=90,shadow=True,explode = (0,0.1),radius = 1.2,autopct='%.1f%%')
plt.legend()
plt.title("Pie Chart")
plt.show()
```

```python
plt.scatter(d['age'],d['pot'])
plt.xlabel("age")
plt.ylabel("pot")
plt.title("Scatter Plot: age vs pot")
plt.show()
```

```python
plt.bar(d['al'],d['sc'], width=0.8,color=['green'])
plt.xlabel("alcohol")
plt.ylabel("sc")
plt.title("Bar Chart: alcohol vs sc")
plt.show()
```
